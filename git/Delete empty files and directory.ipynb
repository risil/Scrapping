```python
import os

def delete_zero_size_files_and_directories(directory_path):
    for root, dirs, files in os.walk(directory_path, topdown=False):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            if os.path.getsize(file_path) == 0:
                os.remove(file_path)
                print(f"Deleted zero-size file: {file_path}")

        for dir_name in dirs:
            dir_path = os.path.join(root, dir_name)
            if not os.listdir(dir_path):  # Check if directory is empty
                os.rmdir(dir_path)
                print(f"Deleted empty directory: {dir_path}")

if __name__ == "__main__":
    directory_to_clean = "F:\\edx\\DOCILE\\NAAC\\ssr_download"

    delete_zero_size_files_and_directories(directory_to_clean)
```
