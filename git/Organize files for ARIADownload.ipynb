```python
import os
import shutil

def rename_and_merge_folders(base_folder, extensions):
    # Traverse the base folder and its subfolders
    for root, dirs, _ in os.walk(base_folder):
        for folder_name in dirs:
            # Check if the folder name ends with any of the specified extensions
            if any(folder_name.lower().endswith(ext) for ext in extensions):
                old_path = os.path.join(root, folder_name)

                # Remove the extension from the folder name
                new_folder_name, _ = os.path.splitext(folder_name)

                # Create the new path with the renamed folder
                new_path = os.path.join(root, new_folder_name)

                # Check if the new folder name already exists
                if os.path.exists(new_path):
                    # Merge the contents of the existing folder
                    merge_folders(old_path, new_path)
                else:
                    # Rename the folder
                    os.rename(old_path, new_path)
                    print(f"Renaming folder: {old_path} to {new_path}")

def merge_folders(source_folder, destination_folder):
    # Merge the contents of the source folder into the destination folder
    for item in os.listdir(source_folder):
        source_item = os.path.join(source_folder, item)
        destination_item = os.path.join(destination_folder, item)

        # If the item is a directory, recursively merge its contents
        if os.path.isdir(source_item):
            merge_folders(source_item, destination_item)
        else:
            # If the item is a file, move it to the destination folder
            shutil.move(source_item, destination_item)
            print(f"Merging file: {source_item} to {destination_item}")

# Replace these values with your actual base folder path and extensions
base_folder_path = 'F:\\edx\\DOCILE\\NAAC\\ARIASSR\\ssr_org'
extensions_to_rename = ['.pdf','.1.pdf', '.xls','.xlsx','.doc', '.docx']

rename_and_merge_folders(base_folder_path, extensions_to_rename)
```

```python
import os

def delete_empty_directories(base_folder):
    # Traverse the base folder and its subfolders in reverse order
    for root, dirs, _ in os.walk(base_folder, topdown=False):
        for dir_name in dirs:
            dir_path = os.path.join(root, dir_name)

            # Check if the directory is empty
            if not os.listdir(dir_path):
                # Delete the empty directory
                print(f"Deleting empty directory: {dir_path}")
                os.rmdir(dir_path)

# Replace this path with your actual base folder path
base_folder_path = 'F:\\edx\\DOCILE\\NAAC\\ARIASSR\\ssr_org'

delete_empty_directories(base_folder_path)
```
