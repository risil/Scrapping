```python
import os
import re
import PyPDF2
from PyPDF2 import PdfReader
# Specify the input folder containing PDF files and the output folder for text files
input_folder = "101-150"
output_folder = "output_texts"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.mkdir(output_folder)

# List all PDF files in the input folder
pdf_files = [f for f in os.listdir(input_folder) if f.endswith(".pdf")]

# Function to extract text from a PDF file and save it as a text file
def convert_pdf_to_text(pdf_file):
    input_path = os.path.join(input_folder, pdf_file)
    output_file = os.path.splitext(pdf_file)[0] + ".txt"
    output_path = os.path.join(output_folder, output_file)
    with open(input_path, "rb") as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        text = ""
        page = pdf_reader.pages[0]
        text = page.extract_text()
        email_addresses = re.findall(r'\S+@\S+', text)
        result_string = ', '.join(email_addresses)
    with open(output_path, "w", encoding="utf-8") as text_file:
        text_file.write(result_string)

       
        
        
        
# Iterate through each PDF file and convert it to text
for pdf_file in pdf_files:
    convert_pdf_to_text(pdf_file)
    print(f"Converted {pdf_file} to text.")

print("Conversion of PDFs to text is complete.")
```
