```python
import os
import shutil
exclude_files = ['aria2c.exe','data.txt','ARIA.txt']
def scan_and_move(source_folder, destination_folder):
    # Create destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Traverse the source folder and its subfolders
    for root, dirs, files in os.walk(source_folder):
        for file in files:
            file_path = os.path.join(root, file)

            # Delete files with zero bytes
            if os.path.getsize(file_path) == 0:
                print(f"Deleting file: {file_path}")
                os.remove(file_path)
            else:
                # Move non-empty files to the destination folder
                destination_path = os.path.join(destination_folder, file)
                #print(f"Moving file: {file_path} to {destination_path}")
                shutil.move(file_path, destination_path)

        # Move empty subfolders to the destination folder
        for dir_name in dirs:
            dir_path = os.path.join(root, dir_name)
            if not os.listdir(dir_path):
                destination_path = os.path.join(destination_folder, dir_name)
                print(f"Moving empty folder: {dir_path} to {destination_path}")
                shutil.move(dir_path, destination_path)

# Replace these paths with your actual source and destination folder paths
source_folder_path = 'E:\\aria'
destination_folder_path = 'F:\\edx\\DOCILE\\NAAC\\All_download_combined'

scan_and_move(source_folder_path, destination_folder_path)
```

```python
def organize_files(source_folder, destination_folder):
    # Create destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Traverse the source folder
    for file_name in os.listdir(source_folder):
        file_path = os.path.join(source_folder, file_name)

        # Check if it's a file
        if os.path.isfile(file_path):
            # Extract characters after the last underscore in the file name
            folder_name, _ = os.path.splitext(file_name)
            folder_name = folder_name.rsplit('_', 1)[-1]

            # Remove any file extension from the folder name
            folder_name = os.path.splitext(folder_name)[0]

            # Create a folder with the extracted name if it doesn't exist
            folder_path = os.path.join(destination_folder, folder_name)
            if not os.path.exists(folder_path):
                os.makedirs(folder_path)

            # Move the file to the corresponding folder
            destination_path = os.path.join(folder_path, file_name)
            print(f"Moving file: {file_path} to {destination_path}")
            shutil.move(file_path, destination_path)

# Replace these paths with your actual source and destination folder paths
source_folder_path = 'F:\\edx\\DOCILE\\NAAC\\All_download_combined'
destination_folder_path = 'F:\\edx\\DOCILE\\NAAC\\ARIASSR\\ssr_org'

organize_files(source_folder_path, destination_folder_path)
```

```python
def rename_and_merge_folders(base_folder, extensions):
    # Traverse the base folder and its subfolders
    for root, dirs, _ in os.walk(base_folder):
        for folder_name in dirs:
            # Check if the folder name ends with any of the specified extensions
            if any(folder_name.lower().endswith(ext) for ext in extensions):
                old_path = os.path.join(root, folder_name)

                # Remove the extension from the folder name
                new_folder_name, _ = os.path.splitext(folder_name)

                # Create the new path with the renamed folder
                new_path = os.path.join(root, new_folder_name)

                # Check if the new folder name already exists
                if os.path.exists(new_path):
                    # Merge the contents of the existing folder
                    merge_folders(old_path, new_path)
                else:
                    # Rename the folder
                    os.rename(old_path, new_path)
                    print(f"Renaming folder: {old_path} to {new_path}")

def merge_folders(source_folder, destination_folder):
    # Merge the contents of the source folder into the destination folder
    for item in os.listdir(source_folder):
        source_item = os.path.join(source_folder, item)
        destination_item = os.path.join(destination_folder, item)

        # If the item is a directory, recursively merge its contents
        if os.path.isdir(source_item):
            merge_folders(source_item, destination_item)
        else:
            # If the item is a file, move it to the destination folder
            shutil.move(source_item, destination_item)
            print(f"Merging file: {source_item} to {destination_item}")

# Replace these values with your actual base folder path and extensions
base_folder_path = 'F:\\edx\\DOCILE\\NAAC\\ARIASSR\\ssr_org'
extensions_to_rename = ['.pdf','.1.pdf', '.xls','.xlsx','.doc', '.docx']

rename_and_merge_folders(base_folder_path, extensions_to_rename)
```
