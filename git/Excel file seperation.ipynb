```python
import os
import shutil
import fnmatch

def copy_matching_excel_files(source_directory, destination_directory):
    # Create the destination directory if it doesn't exist
    if not os.path.exists(destination_directory):
        os.makedirs(destination_directory)

    # Iterate through all files in the source directory
    for root, dirs, files in os.walk(source_directory):
        for file_name in fnmatch.filter(files, '*.xlsx'):  # Change the pattern if needed
            # Check if the file name matches the pattern
            if '2.4.1' in file_name or '2.7.1' in file_name or 'dynamic' in file_name:
                source_path = os.path.join(root, file_name)
                destination_path = os.path.join(destination_directory, file_name)

                # Copy the file to the destination directory
                shutil.copy2(source_path, destination_path)
                print(f"Copied: {source_path} to {destination_path}")

# Example usage
source_directory = "F:\\edx\\DOCILE\\NAAC\\ssr_download"
destination_directory = "D:/Myxls"

copy_matching_excel_files(source_directory, destination_directory)
```
