```python
import re 
import os
import pdfminer
import pdfminer.high_level
from pdfminer.high_level import extract_text 

pdf_folder = 'D:\\DOCILE\\Atal_FDP'
def extract_emails(pdf_path): 
    emails = set() 
    for page_num in range(1, extract_text(pdf_path, page_numbers=[i for i in range(1, 1000)]).count('\n')): 
        text = extract_text(pdf_path, page_numbers=[page_num]) 
        matches = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b', text) 
        for match in matches: 
            emails.add(match) 
    return emails
```

```python
# List all PDF files in the folder
pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]
mcolln = set() 
# Extract and print email addresses from each PDF
for pdf_file in pdf_files:
    pdf_path = os.path.join(pdf_folder, pdf_file)
    print(f'{pdf_file}:')
    emails = extract_emails(pdf_path)
    mcolln = "\n".join(emails)
    if emails:
        print(f'{pdf_file}:')
        for email in emails:
            print(email)
            
file1=open(r"D:\\DOCILE\\mails.txt","a")
file1.writelines(mcolln)
```
