```python
import fitz  # PyMuPDF
import re
import os

# Specify the folder containing the PDF files
pdf_folder = 'E:\\NAAC\\SAR\\Maharastra'

# Regular expression for matching email addresses
regex= r'[A-Za-z0-9\s]+ COLLEGE [A-Za-z0-9\s]+'
email_regex = r'AISHE [A-Za-z]+\s+:\s+[A-Za-z]-[A-Za-z0-9]+'
trackID=r'Institution Track ID\s+:\s+[A-Za-z0-9]+'
trck2=r'AISHE [A-Za-z]+\s+:\s+[A-Za-z]-[A-Za-z0-9]+'

# Function to extract email addresses from a PDF file
def extract_emails_from_pdf(pdf_file):
    doc = fitz.open(pdf_file)
    text = ''
    
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        text += page.get_text()

    emails = re.findall(regex, text)
    return emails

# List all PDF files in the folder
pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]

# Extract and print email addresses from each PDF
for pdf_file in pdf_files:
    pdf_path = os.path.join(pdf_folder, pdf_file)
    email_addresses = extract_emails_from_pdf(pdf_path)
    mcolln = "\s".join(email_addresses)
    if email_addresses:
        print(f'{pdf_file}:')
        mcoll = "\n".join(email_addresses)
        for email in email_addresses:
            print(email)

# You can also save the email addresses to a file if needed.
#file1=open(r"D:\\DOCILE\\dmails.txt","a")
#file1.writelines(mcolln)
file_path = "iiqa.txt"
with open(file_path, "w") as file:
    file.write(mcoll)
```
