```python
import fitz  # PyMuPDF
import re
import os

def extract_doi_from_pdf(file_path):
    try:
        doc = fitz.open(file_path)
        dois = set()  # Using a set to store unique DOIs

        for page_num in range(doc.page_count):
            page = doc.load_page(page_num)
            text = page.get_text("text")
            
            # Regular expression to find DOIs
            doi_pattern = r"\b10\.\d{4,}(?:\.\d+)*\/\S+(?![^<]*>)"
            
            dois.update(re.findall(doi_pattern, text))
        print(doc)
        print(dois)   
        doc.close()
        return dois

    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return set()

def extract_dois_from_folder(folder_path):
    dois_found = set()

    for file in os.listdir(folder_path):
        if file.lower().endswith('.pdf'):
            file_path = os.path.join(folder_path, file)
            dois = extract_doi_from_pdf(file_path)
            dois_found.update(dois)

    return dois_found

# Replace this path with the folder containing your PDF files
pdf_folder_path = 'D:\DOCILE\Sensor_Journal\pending'

extracted_dois = extract_dois_from_folder(pdf_folder_path)

if extracted_dois:
    print("DOIs found:")
    for doi in extracted_dois:
        print(doi)
else:
    print("No DOIs found in the PDF files.")
```
