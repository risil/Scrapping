```python
import os
import openpyxl

def find_emails_in_sheet(sheet):
    email_entries = []
    for row in sheet.iter_rows(min_row=2, values_only=True):
        for cell_value in row:
            if isinstance(cell_value, str) and "@" in cell_value:
                email_entries.append(", ".join(str(cell) for cell in row))
                
                break  # Move to the next row after finding the first email in the current row
    return email_entries

def process_excel_file(file_path, output_file_path):
    try:
        wb = openpyxl.load_workbook(file_path, read_only=True)
        with open(output_file_path, 'a', encoding='utf-8') as output_file:
            for sheet_name in wb.sheetnames:
                sheet = wb[sheet_name]
                email_entries = find_emails_in_sheet(sheet)
                output_file.write('\n'.join(email_entries) + '\n')
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

def scan_and_process_excel_files(source_folder, output_file_path):
    if os.path.exists(output_file_path):
        os.remove(output_file_path)

    for foldername, subfolders, filenames in os.walk(source_folder):
        for filename in filenames:
            if filename.lower().endswith(('.xls', '.xlsx')):
                file_path = os.path.join(foldername, filename)
                process_excel_file(file_path, output_file_path)

if __name__ == "__main__":
    source_directory = "F:\\edx\\DOCILE\\NAAC\\ssr_download"
    output_file_path = "D:\\DOCILE\\Final Mail ID collection\\NAACmailname_1201.txt"

    scan_and_process_excel_files(source_directory, output_file_path)
```

```python
import os
import openpyxl
import xlrd

def find_emails_in_sheet_xlsx(sheet):
    email_entries = []
    for row in sheet.iter_rows(min_row=2, values_only=True):
        for cell_value in row:
            if isinstance(cell_value, str) and "@" in cell_value:
                email_entries.append(", ".join(str(cell) for cell in row))
                break  # Move to the next row after finding the first email in the current row
    return email_entries

def find_emails_in_sheet_xls(sheet):
    email_entries = []
    for row_num in range(1, sheet.nrows):
        row = sheet.row_values(row_num)
        for cell_value in row:
            if isinstance(cell_value, str) and "@" in cell_value:
                email_entries.append(", ".join(str(cell) for cell in row))
                break  # Move to the next row after finding the first email in the current row
    return email_entries

def process_excel_file(file_path, output_file_path):
    try:
        if file_path.lower().endswith('.xlsx'):
            wb = openpyxl.load_workbook(file_path, read_only=True)
            for sheet_name in wb.sheetnames:
                sheet = wb[sheet_name]
                email_entries = find_emails_in_sheet_xlsx(sheet)
                with open(output_file_path, 'a', encoding='utf-8') as output_file:
                    output_file.write('\n'.join(email_entries) + '\n')
        elif file_path.lower().endswith('.xls'):
            workbook = xlrd.open_workbook(file_path)
            for sheet_name in workbook.sheet_names():
                sheet = workbook.sheet_by_name(sheet_name)
                email_entries = find_emails_in_sheet_xls(sheet)
                with open(output_file_path, 'a', encoding='utf-8') as output_file:
                    output_file.write('\n'.join(email_entries) + '\n')
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

def scan_and_process_excel_files(source_folder, output_file_path):
    if os.path.exists(output_file_path):
        os.remove(output_file_path)

    for foldername, subfolders, filenames in os.walk(source_folder):
        for filename in filenames:
            file_path = os.path.join(foldername, filename)
            process_excel_file(file_path, output_file_path)

if __name__ == "__main__":
    source_directory = "D:\\Myxls"
    output_file_path = "D:\\DOCILE\\Final Mail ID collection\\NAACmailname_1501.txt"

    scan_and_process_excel_files(source_directory, output_file_path)
```
