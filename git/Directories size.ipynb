```python
import os

def get_directory_size(directory):
    total_size = 0

    # Walk through all directories and files
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            total_size += os.path.getsize(file_path)

    return total_size

def display_directory_sizes(base_directory):
    print(f"Directory Sizes in {base_directory}:")
    
    # Walk through all directories in the base directory
    for root, dirs, _ in os.walk(base_directory):
        for directory in dirs:
            directory_path = os.path.join(root, directory)
            size = get_directory_size(directory_path)/1000000000
            print(f"{directory_path}: {size} GB")

# Example usage
base_directory_to_search = "F:\\edx\\DOCILE\\NAAC"
display_directory_sizes(base_directory_to_search)
```
