```python
import os
import subprocess

def download_file(url, folder_path, aria2c_path):
    file_name = url.split("/")[-1]
    file_path = os.path.join(folder_path, file_name)
    
    command = [aria2c_path, url, "-d", folder_path, "-o", file_name]
    
    try:
        subprocess.run(command, check=True)
        print(f"Downloaded {file_name} and saved in {folder_path}")
    except subprocess.CalledProcessError as e:
        print(f"Failed to download {url}: {e}")

def process_urls(file_path, aria2c_path):
    with open(file_path, 'r') as file:
        urls = file.read().splitlines()

    for url in urls:
        folder_name = url.split("/")[-2]
        folder_path = os.path.join('E:\\NAAC\\All_downloaded', folder_name)

        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

        download_file(url, folder_path, aria2c_path)

if __name__ == "__main__":
    input_file_path = "C:\\aria\\data.txt"  # Replace with your actual input file path
    aria2c_path = r"C:\aria\aria2c.exe"  # Replace with the actual path to your aria2c.exe
    
    process_urls(input_file_path, aria2c_path)
```
