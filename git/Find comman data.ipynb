```python
import pandas as pd
import glob

# Function to read and accumulate data from Excel sheets
def read_excel_sheets(file_path):
    xls = pd.ExcelFile(file_path)
    all_data = pd.DataFrame()
    for sheet_name in xls.sheet_names:
        sheet_data = xls.parse(sheet_name, header=None)
        all_data = pd.concat([all_data, sheet_data], ignore_index=True)
    return all_data

# Function to read and accumulate data from text files
def read_text_files(file_pattern):
    all_data = pd.DataFrame()
    for file_path in glob.glob(file_pattern):
        text_data = pd.read_csv(file_path, header=None)
        all_data = pd.concat([all_data, text_data], ignore_index=True)
    return all_data

# Function to perform union of two datasets on the common item
def union_datasets(data1, data2, common_column):
    merged_data = pd.merge(data1, data2, how='outer', left_on=common_column, right_on=common_column)
    return merged_data

# Example usage
excel_file_path = "C:\\Users\\Admin\\Desktop\\AICTE.xlsx"
text_files_pattern = 'D:\\DOCILE\\Final Mail ID collection\\AICTE Text files\\*.txt'
single_text_file_path = 'D:\\DOCILE\\Final Mail ID collection\\AICTE Text files\\1501.00349.txt'

excel_data = read_excel_sheets(excel_file_path)
excel_data.to_csv('D:\DOCILE\Final Mail ID collection\AICTE Text files\excel_data.csv', index=False)
```

```python
def read_single_text_file(file_path):
    text_data = pd.read_csv(file_path, header=None,error_bad_lines=False)
    return text_data

text_data = read_single_text_file(single_text_file_path)
#text_data = read_text_files(text_files_pattern)

common_column_name = 0  # Assuming the common column is the first column (adjust as needed)

merged_data = union_datasets(excel_data, text_data, common_column_name)

# Display the final merged data
print(merged_data)
merged_data.to_csv('D:\DOCILE\Final Mail ID collection\AICTE Text files\merged_data.csv', index=False)
```
